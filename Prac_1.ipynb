```python
# Step Function
def step(inputs_sum,threshold):
  if inputs_sum>= threshold:
    return 1
  else:
    return 0
```

```python
# OR GATE
def or_an(inputs) :
  inputs_sum = 0
  threshold = 1
  for i in inputs:
    inputs_sum += i
  return step(inputs_sum, threshold)
```

```python
inputs = [[0,0],[0,1],[1,0],[1, 1]]

for i in inputs:
  print(i, or_an(i))
```

```python
# AND GATE
def and_an(inputs) :
  inputs_sum = 0
  threshold = 2
  for i in inputs:
    inputs_sum += i
  return step(inputs_sum, threshold)
```

```python
inputs = [[0,0],[0,1],[1,0],[1, 1]]

for i in inputs:
  print(i, and_an(i))
```

```python
# NOT Gate
def not_an(x):
    weight = -1
    threshold = 0

    inputs_sum = x * weight
    return step(inputs_sum, threshold)

# Testing
print(not_an(0))
print(not_an(1))
```

```python
# Implementing NAND using MCP
def nand_an(inputs):
  and_result = and_an(inputs)
  return not_an(and_result)
```

```python
inputs = [[0,0],[0,1],[1,0],[1,1]]
print("NAND Gate Output:")
for i in inputs:
  print(i, nand_an(i))
```

```python
# Implementing NOR using MCP
def nor_an(inputs):
  or_result = or_an(inputs)
  return not_an(or_result)
```

### Gates Using Perceptron

```python
# OR Using Perceptron
def and_func(inputs,weights):
  inputs_sum=0
  threshold = 0
  bias = -1
  for i in range(len(inputs)):
    inputs_sum += inputs[i]*weights[i]
  return step(inputs_sum+bias,threshold)
```

```python
inputs = [[0,0],[0,1],[1,0],[1,1]]
print("OR Gate Output:")
for i in inputs:
  weight = [1,1]
  output = and_func(i,weight)
  print(i,output)
```

```python
# AND Using Perceptron
def and_func(inputs,weights):
  inputs_sum=0
  threshold = 0
  bias = -2
  for i in range(len(inputs)):
    inputs_sum += inputs[i]*weights[i]
  return step(inputs_sum+bias,threshold)
```

```python
inputs = [[0,0],[0,1],[1,0],[1,1]]
print("AND Gate Output:")
for i in inputs:
  weight = [1,1]
  output = and_func(i,weight)
  print(i,output)
```

```python
# NOT Using Perceptron
def not_func(inputs, weights):
    inputs_sum = 0
    threshold = 0
    bias = 0

    for i in range(len(inputs)):
        inputs_sum += inputs[i] * weights[i]

    return step(inputs_sum + bias, threshold)
```

```python
inputs = [[0], [1]]
weights = [-1]

print("NOT Gate Output:")
for i in inputs:
    print(i, not_func(i, weights))
```

```python
# NAND Gate Using Perceptron
def nand_func(inputs):
    # AND gate first
    and_weights = [1, 1]
    and_output = and_func(inputs, and_weights)

    # NOT gate on AND output
    not_weights = [-1]
    nand_output = not_func([and_output], not_weights)

    return nand_output
```
