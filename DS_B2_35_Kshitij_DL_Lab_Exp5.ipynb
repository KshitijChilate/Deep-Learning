### Aim : Use MNIST dataset to train neural networks . Experiments with Architecture of neural networks, and with different activation functions for hidden and output layers. Also work with various optimizers.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import mnist
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
print("x_train shape:", X_train.shape)
print(X_train.shape[0], "train samples")
print(X_test.shape[0], "test samples")
```

```python
def create_model(hidden_layers = 1, neurons = 64, activation = 'relu',input_shape = (28,28)):
  #input
  model = Sequential()
  model.add(Flatten(input_shape = input_shape))
  #hidden layer
  for _ in range(hidden_layers):
    model.add(Dense(neurons, activation = activation))
  #Output layer
  model.add(Dense(10, activation='softmax'))
  return model
```

### Model - 1 (using sgd) (activation - relu and output layer = softmax)

```python
model = create_model(hidden_layers= 2,neurons = 64, activation = 'relu',input_shape = (28,28))
model.compile(optimizer = 'sgd',loss = 'sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_split= 0.1)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test accuracy:', test_acc)
```
